```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv('balance.txt', delim_whitespace=True)
```

```python
# compare the average income based on ethnicity 
print(df.groupby('Ethnicity')['Income'].mean(numeric_only=True))
```

```python
#On Average , do married or single people have a higher balance ?
print("The average income of married people is", df[df.Married == "Yes"].loc[:,"Balance"].mean())
print("The average income of single people is ", df[df.Married == "No"].loc[:,"Balance"].mean())
```

```python
#What is the highest income in our dataset ?
print("The highest income in our dataste is", df.loc[:, "Income"].max())
```

```python
#What is the lowest income in our dataset ?
print("The lowest income in our dataste is", df.loc[:, "Income"].min())
```

```python
#How many Cards do we have recorded in our dataset?
print("We have this amount of cards in our dataset", df.loc[:, "Cards"].sum())
```

```python
#How many females do we have information for vs how many males ?
print(df.groupby('Gender')['Gender'].count())
```
